# Correción y Filtración

```python
# Quitar comas y convertir a número entero
df_ESTABLECIMIENTOS_EDU['año'] = df_ESTABLECIMIENTOS_EDU['año'].astype(str).str.replace(',', '').astype(int)


print(df_ESTABLECIMIENTOS_EDU['año'].value_counts())
```

```python
# Extraer el número y convertirlo a entero
df_ESTABLECIMIENTOS_EDU['estrato_Socio_Economico '] = (
    df_ESTABLECIMIENTOS_EDU['estrato_Socio_Economico']
    .astype(str)                     # Asegura que los valores sean texto
    .str.extract(r'(\d+)')           # Extrae el número (como texto)
    .astype('Int64')                 # Convierte a entero, permitiendo NaN
)


print(df_ESTABLECIMIENTOS_EDU['estrato_Socio_Economico'])
```

```python
df_ESTABLECIMIENTOS_EDU['estrato_Socio_Economico'] = (
    df_ESTABLECIMIENTOS_EDU['estrato_Socio_Economico']
    .astype(str)                     # Ensure values are strings
    .str.extract(r'(\d+)')           # Extract the digit(s)
    .fillna('0')                      # Fill NaN values with '0' string
    .astype(int)                     # Convert to integer
)

print(df_ESTABLECIMIENTOS_EDU['estrato_Socio_Economico'].value_counts())
```

```python
# 1. Asegúrar de que no haya valores nulos
df_ESTABLECIMIENTOS_EDU['niveles'] = df_ESTABLECIMIENTOS_EDU['niveles'].fillna('SIN DATO')

# 2. Divir por comas y crea filas nuevas
df_niveles = df_ESTABLECIMIENTOS_EDU.assign(nivel_individual=df_ESTABLECIMIENTOS_EDU['niveles'].str.split(',')).explode('nivel_individual')

# 3. Limpiar espacios en blanco
df_niveles['nivel_individual'] = df_niveles['nivel_individual'].str.strip()

# 4. Elimina columna original si ya no la necesitas
df_niveles = df_niveles.drop(columns=['niveles'])

print(df_niveles)
```

```python
# Filtrar solo los establecimientos del departamento de Atlantico
# Para evitar problemas con nombres en mayúscula/minúscula
df_ESTABLECIMIENTOS_EDU['Departamentoatlántico'] = df_ESTABLECIMIENTOS_EDU['nombredepartamento'].str.upper()
df_ATLÁNTICO = df_ESTABLECIMIENTOS_EDU[df_ESTABLECIMIENTOS_EDU['nombredepartamento'] == 'ATLÁNTICO']

display(df_ATLÁNTICO)

```
